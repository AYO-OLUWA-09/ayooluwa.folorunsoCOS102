```python
# getting the workers age and experience
age = int(input("Enter Age:"))
exp = int(input("Enter Experience: "))

# conditions for the ATR

if exp > 25 and age >= 55:
    atr = 5600000
elif exp > 20 and age >= 45:
    atr = 4480000
elif exp > 10 and age >= 35:
    atr = 1500000
else:
    atr = 550000

#print the atr
print("Annual Tax Revenue: N", atr)
```
